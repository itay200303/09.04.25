```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
years = [2015, 2016, 2017, 2018, 2019]
height = [110, 115, 120, 125, 132]

plt.plot(years, height, marker='o', linestyle='-', color='blue')
plt.title('height over years')

plt.xticks(years)
plt.xlabel('year')
plt.ylabel('height in cm')

plt.grid(True)
plt.show()
```

```python
months = ["Jan", "Feb", "Mar", "Apr", "May"]
temperature = [12, 14, 17, 21, 25]
rain = [80, 60, 55, 40, 20]

fig, axs = plt.subplots(1, 2, figsize=(12, 5))

axs[0].plot(months, temperature, marker='o', color='red')
axs[0].set_title('temperature in months')
axs[0].set_xlabel('month')
axs[0].set_ylabel('temperature (°C)')
axs[0].grid(True)

axs[1].bar(months, rain, color='blue')
axs[1].set_title('amount of rain in months')
axs[1].set_xlabel('month')
axs[1].set_ylabel('rain(mm)')
axs[1].grid(True, axis='y')

plt.tight_layout()
plt.show()
```

```python
grades = [70, 85, 90, 76, 65, 95, 88, 73, 80, 78, 67, 92, 60, 83, 89, 55, 91, 74, 69, 87, 81, 79, 62, 77, 93]

fig, axs = plt.subplots(1, 2, figsize=(12, 5))

axs[0].hist(grades, bins=5, color='skyblue', edgecolor='black')
axs[0].set_title('hist grades - 5 groups')
axs[0].set_xlabel('grade')
axs[0].set_ylabel('students')

axs[1].hist(grades, bins=10, color='lightyellow', edgecolor='black')
axs[1].set_title('hist grades - 10 groups')
axs[1].set_xlabel('grade')
axs[1].set_ylabel('students')

plt.tight_layout()
plt.show()
```

```python
age_groups = ['10-12', '13-15', '16-18', '19-21']
students_count = [30, 45, 25, 10]

plt.bar(age_groups, students_count, color='cyan')
plt.title('comparison between students_count to age_group')
plt.xlabel('age_group')
plt.ylabel('students_count') 

plt.grid(True, axis='y')
plt.show()
```
